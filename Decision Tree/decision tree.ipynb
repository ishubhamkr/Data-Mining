```python
#Load libraries
import pandas as pd
from sklearn import *
from sklearn.tree import export_graphviz
from sklearn.metrics import *
```

```python
# load dataset
```

```python
df=pd.read_csv('hcvdat0.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
#Check for NULL Values
df[df.isnull().any(axis=1)].head()
```

```python
# Drop NULL Column
df=df.dropna()
df.shape
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
category=LabelEncoder()
sex=LabelEncoder()
df['category_df']=category.fit_transform(df['Category'])
df['sex_df']=category.fit_transform(df['Sex'])
df.head()
```

```python
# Copy data set of category into data_class
data_class=df['Category'].copy()
data_class=list(dict.fromkeys(data_class))
print(data_class)
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = df.drop(['Category','Sex'], axis=1)
y=df['category_df'].copy()
X = df.drop(['category_df'], axis=1)
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.3,random_state=500)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_leaf_nodes = 15, random_state = 0 )
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_predicted = model.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy : ", accuracy_score(y_test,y_predicted)*100)
print("Report : ", classification_report(y_test,y_predicted))
print(confusion_matrix(y_test,y_predicted))
```

```python
dot_data = export_graphviz( model,out_file ='tree.dot',
                           feature_names = X_train.columns, 
                           class_names =data_class,filled = True,
                           rounded = True,special_characters = True)
```
